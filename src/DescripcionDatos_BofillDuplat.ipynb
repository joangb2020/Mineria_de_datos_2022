1. Descripción de la base de datos.

La base de datos fue descargada de https://www.datos.gov.co/widgets/ynam-yc42, a partir de la información suministrada por el Instituto Colombiano para la Evaluación de la Educación - Icfes. Consta de la información de los evaluados durante el examen Saber 11 de la segunda aplicación de 2019. Contiene información des cuestionario socioeconómico, los puntajes de las 5 pruebas e información básica del colegio. En total son 546212 evaluados y 82 variables. 

Dado que son tantas variables, describiremos a continuación solamente algunas de ellas. Las variables que  contienen el nombre de una prueba (ej: Matemáticas) están en la base y representan a las demás, existen las 5 pruebas que componen el examen (Matemáticas, Lectura, Sociales y Ciudadanas, Inglés, Ciencias). Por otro lado, variables como códigos Dane de la sede, del colegio o de jurisdicciones territoriales serán omitidas pues no son de interés para el proyecto.

**Nominales**

'ESTU_TIPODOCUMENTO': tipo de documento, puede tomar valores TI, CC, Pasaporte, ...

'ESTU_NACIONALIDAD': Colombia, Venezuela, ...

'ESTU_GENERO', Masculino, Femenino, Otro...

'ESTU_CONSECUTIVO': identificador único del evaluado.

**Ordinales**

 'FAMI_EDUCACIONPADRE': Nivel máximo de educación del padre, toma valores como Primaria, media secundaria, pregrado, ...

 'FAMI_EDUCACIONMADRE': Nivel máximo de educación de la madre.

 'DESEMP_MATEMATICAS': Nivel de desempeño para la prueba de matemáticas del evaluado, puede tomar los valores 1, 2, 3 y 4, siendo este último el nivel máximo.

 'FAMI_ESTRATOVIVIENDA': estrato del hogar del evaluado. 

**Intervalo o escala**

'ESTU_FECHANACIMIENTO': fecha de nacimiento del evaluado.

'PERCENTIL_C_NATURALES': percentil en que se encuentra el evaluado con respecto a toda la población en la prueba de Ciencias Naturales.

**Razón**

'ESTU_DEDICACIONLECTURADIARIA', Horas al día que dedica a leer el evaluado.

'ESTU_HORASSEMANATRABAJA', horas a la semana que trabaja el evaluado.

'PUNT_C_NATURALES', puntaje en escala 0-100 en la prueba de Ciencias Naturales.
